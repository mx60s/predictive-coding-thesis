```python
def remove_repeats(coords, frames):
    new_coords = []
    new_frames = []
    for i in range(1, len(coords)):
        if abs(coords[i][0] - coords[i-1][0]) + abs(coords[i][1] - coords[i-1][1]) == 1:
            new_coords.append(coords[i])
            new_frames.append(frames[i])

    return new_coords, new_frames
```

```python
coords = np.load('data/face-forward/' + 'coords_random_worldone_2023-12-28-20-27-29.npy').astype(int)
frames = np.load('data/face-forward/' + 'frames_random_worldone_2023-12-28-20-27-29.npy').astype(int)

n_coords, n_frames = remove_repeats(coords, frames)

#n_coords = n_coords[30:]
#n_frames = n_frames[30:]

print(len(coords))
print(len(n_coords))

fig, axes = plt.subplots(1, 8, figsize=(8, 2))

for i in range(8):
    axes[i].imshow(n_frames[i])
    axes[i].axis('off')

plt.tight_layout()
plt.show()

print(n_coords[:8])
```
